# Bivariate Analysis

# 5. Is there a correlation between sales_price and the sales ?

```python
# Convert 'sales' column to numerical values
dataset['sales'] = dataset['sales'].map({'Yes': 1, 'No': 0})

# Now compute the correlation coefficient
correlation = dataset['sales_price'].corr(dataset['sales'])

print("Correlation Coefficient:", correlation)
```

# 6. Does battery_capacity correlate with the sales_price ?

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(dataset['battery_capacity'], dataset['sales_price'], alpha=0.5)
plt.title('battery_capacity vs sales_price')
plt.xlabel('battery_capacity')
plt.ylabel('sales_price')
plt.grid(True)
plt.show()
```

# 7. Is there an association between RAM and the need for ratings?

```python
# Create a contingency table
contingency_table = pd.crosstab(dataset['RAM'], dataset['ratings'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the chi-square statistic and p-value
print("Chi-square Statistic:", chi2)
print("p-value:", p)
```

# Correlation Heatmap of Tumor Characteristics

```python
dataset.columns
```

```python

# Select relevant columns for heatmap
relevant_columns = ['ROM',
       'RAM', 'display_size', 'num_rear_camera', 'battery_capacity', 'ratings',
       'sales_price', 'discount_percent', 'sales']

# Calculate correlation matrix
correlation_matrix = dataset[relevant_columns].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Tumor Characteristics')
plt.show()

```

# Covariance Heatmap of Tumor Characteristics

```python

# Select relevant columns for heatmap
relevant_columns = ['ROM',
       'RAM', 'display_size', 'num_rear_camera', 'battery_capacity', 'ratings',
       'sales_price', 'discount_percent', 'sales']

# Calculate covariance matrix
covariance_matrix = dataset[relevant_columns].cov()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(covariance_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Covariance Heatmap of Tumor Characteristics')
plt.show()
```
